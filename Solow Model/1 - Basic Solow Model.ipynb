### Modelo de Solow-Swan:

###### Con los siguientes parámetros, encuentren y grafiquen los valores de estado estacionario de Kt, It, Ct y ∆k(t):
*Tasa de depreciacion del capital   =     0.1*

*Tasa de participacion del capital   =   0.5*

*Tasa de crecimiento de la poblacion  =   0*

*Tasa de ahorro            =              0.3*

*Nivel de productividad        =          1*

*Capital inicial              =           4*

*Poblacion inicial.           =           1*

```python
import numpy as np               
import matplotlib.pyplot as plt  
```

```python
# Parameters

K_size = 30                      # Our model test size / domain = 30 (for example)
A = 1                            # Technology/Productivity = 1
N = 1                            # Labour (poblacion inicial) = 1
n = 0                            # Poblation growth rate = null
alpha = 0.50                     # K share (Tasa de participación del capital) = 0.5
s = 0.30                         # Savings rate = 0.3
d = 0.10                         # K Depreciation rate = 0.1 

#Arrays
K = np.arange(K_size)            # Create empty array of K (size=K_size=30)

print(K)                         # Check array capital (de 0 a 29)
```

```python
# Functions

def output(K):   # Cobb-Douglas Production Function
    Y = A * (K)**(alpha) * (N)**(1-alpha)    
    return Y
```

```python
# Arrays que nos faltan (notar que Y será de tamaño K.shape) y por ende tendremos todos arrays de igual dimension
Y = output(K)
D = d*K
I = s*Y

# Check
print(Y)
print(D)
print(I)
```

```python
# Steady State (Star) Values

Kstar = ((s*A)/(d+n))**( 1/(1-alpha) )
Ystar = A*(Kstar**alpha)
Istar = s*Ystar
Cstar = Ystar - Istar
Dstar = d*Kstar

print("La cantidad de capital de equilibrio es: ", np.around(Kstar, decimals=1))
print("El nivel de producto de equilibrio es: ", np.around(Ystar, decimals=1))
print("El nivel de inversion de equilibrio es: ", np.around(Istar, decimals=1))
print("El nivel de consumo de equilibrio es: ", np.around(Cstar, decimals=1))
print("El nivel de depreciación de equilibrio es: ", np.around(Dstar, decimals=1))
```

```python
# Graphics - Plot the Solow Model

y_max = np.max(Y)              # Usaremos luego el valor para delimitar el tamaño del eje Y

fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(K, Y, "k", ls = '-', label="Output")        # Eje X: capital / Eje Y: Ingreso
ax.plot(K, I, "b", ls = '-', label="Investment")    # Eje X: capital / Eje Y: Inversion
ax.plot(K, D, "r", ls = '-', label="Depreciation")  # Eje X: capital / Eje Y: Depreciacion

ax.set(title="Solow Model", xlabel="Capital Stock")   # Seteamos titulo y etiqueta para el eje X
plt.legend(loc=2)    # Location de las leyendas o labels para cada función (esquina izquierda)

plt.axvline(x = Kstar, ls = ":", color = 'k')  # Linea (punteada) en el valor de equilibrio K
plt.axhline(y = Istar, ls = ":", color = 'k')  # Linea (punteada) en el valor de equilibrio I
plt.axhline(y = Ystar, ls = ":", color = 'k')  # Linea (punteada) en el valor de equilibrio Y

plt.axis([0, K_size, 0, y_max])    # Seteamos los ejes a partir de los datos de una lista que tiene X=(0,K_size) ; Y = (0,y_max)

plt.show()
```
