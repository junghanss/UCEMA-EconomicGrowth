# Tipos de datos básicos
Hay cuatro tipos básicos de datos

- $\textbf{int}$ enteros

- $\textbf{float}$ punto flotantes (números decimales)

- $\textbf{Bool}$ boolean (True o False)

- $\textbf{String}$ texto 

```python
x=10 #interger
y=15.2 #Float
z='Hello World' #String
xx=int(10)
yy=float(15.2)
zz=str(2)
xy=int(2.9)
zx = True
a,b,c = 'Arg','Chl','Br'
```

```python
print(type(x))
print(type(y))
print(type(z))
print(type(zz))
print(a)
print(b)
print(c)
print(zx)
```

```python
str = 'Hello World!'

print(str)          # Muestra la variable entera
print(str[0])       # Muestra la primera letra
print(str[2:5])     # Muestra de la letra en la segunda posición hasta la cuarta posición
print(str[2:])      # Muestra de la letra en la segunda posición hasta el final
print(str * 2)      # Multiplicamos el string dos veces
print(str + "TEST") # Le sumamos la palabra TEST
```

# Tipos de colecciones (recordatorio)
### Los tipos de colecciones más utilizados son:

- $\textbf{Lista}$: una lista ordenada y mutable de valores

- $\textbf{tuple}$: una lista ordenada e inmutable de valores

- $\textbf{set}$: una lista mutable pero no ordenada de valores

- $\textbf{dict}$: un diccionario no ordenado 

```python
mylist = ['matem', 'física', 'química']
mytuple = ('matem', 'física', 'química')
myset = {'matem', 'física', 'química'}
mydic = {'matem':'loglin', 'física':'gravedad', 'química':'compuesto'}
print(mylist)
print(mytuple)
print(myset)
```
