### Modelo de Malthus

#### 1. Replicar los impulso respuesta obtenidos en el handout:

##### Choose values for the variables: z0 = 100; theta = 0.4; Beta = 0.045; and alfa = 0.5. Suppose that the initial population is p0 = 100. We start the economy in a stationary state, where g (ct) = 1, and let it run for four periods. In period 5, we let the technology go up to 120 and let the economy run to period 30.

```python
import numpy as np               
import matplotlib.pyplot as plt  
```

```python
## Creo las variables

theta = 0.4
beta = 0.045
alpha = 0.5
N_i = 100
z_i=100.0
z_f=120.0
#periods =

## Calculo los valores de consumo y tierra per capita en el SS, y la cantidad fija de tierra:

c_star = round(((1/beta)**(1/alpha)),2)

l_star = round(((c_star/z_i)**(1/theta)),2)


tierra = N_i*l_star

#time = np.arange(t_periods) 
##Valores
ct_values =[c_star,c_star,c_star,c_star,c_star]
nt_values =[N_i,N_i,N_i,N_i,N_i]
lt_values =[l_star,l_star,l_star,l_star,l_star]
gt_values =[1,1,1,1,1]

           
##Funciones
def nt(tierra,z_f,theta,beta,alpha):
    nt = 0
    lt=0
    ct=0
    gt=0
    period=0
    while period <=25:
        nt = round((nt_values[-1]*gt_values[-1]),2)
        nt_values.append(nt)
        
        lt = round((tierra/nt_values[-1]),2)  
        lt_values.append(lt)
        
        ct = round((z_f*(lt_values[-1]**theta)),2)
        ct_values.append(ct)
                        
        gt = round((beta*(ct_values[-1]**alpha)),2)
        gt_values.append(gt)
        
        period+=1
    
    return

  
nt(tierra, z_f, theta, beta, alpha)



##Tabla


print('C* = ' + str(round(c_star,2)), '\n','l* = ' + str(round(l_star,2)), '\n')


import pandas as pd
tabla = pd.DataFrame({'Tierra' : tierra , 'Poblacion' : nt_values, 'Consumo' : ct_values, 'Tierra p/ capita' : lt_values, 'Crecimiento' : gt_values})
print(tabla)
##Graficos
#ct
lista=ct_values
v = [0, len(lista), 0.9*min(lista), 1.1*max(lista)]

fig, ax = plt.subplots(figsize=(10, 3))

ax.plot(ct_values, "k", ls = '-', label="C t")

ax.set(title="Modelo de Malthus", xlabel="Consumo per capita")
plt.legend(loc=2)
plt.axis(v)
plt.show()

#lt
lista=lt_values
v = [0, len(lista), 0.9*min(lista), 1.1*max(lista)]

fig, ax = plt.subplots(figsize=(10, 3))

ax.plot(lt_values, "g", ls = '-', label="L t")

ax.set(title="Modelo de Malthus", xlabel="Tierra per capita")
plt.legend(loc=2)
plt.axis(v)
plt.show()


#gt
lista=gt_values
v = [0, len(lista), 0.9*min(lista), 1.1*max(lista)]

fig, ax = plt.subplots(figsize=(10, 3))

ax.plot(gt_values, "r", ls = '-', label="G t")

ax.set(title="Modelo de Malthus", xlabel="Crecimiento")
plt.legend(loc=2)
plt.axis(v)
plt.show()


#ct
lista=nt_values
v = [0, len(lista), 0.9*min(lista), 1.1*max(lista)]

fig, ax = plt.subplots(figsize=(10, 3))

ax.plot(nt_values, "b", ls = '-', label="N t")

ax.set(title="Modelo de Malthus", xlabel="Poblacion")
plt.legend(loc=2)
plt.axis(v)
plt.show()
```
